```python
import pandas as pd
import numpy as np
import json
import random
import csv
import string
import base64
import secrets
import string
import bcrypt
```

```python
data= ""
dataJs = []
with open('Dealer.csv', 'r') as file:
    # Read the CSV data
    csv_data = list(csv.reader(file))

    # Remove the header row if present
    header = csv_data[0]
    data = csv_data[1:]
    
   

data_vehicle = open('vehiclev3.json' , 'r')
data1 = json.load(data_vehicle)
    

def get_vhicleList():
    # Create a list to store chosen JSON objects
    chosen_objects = []

    # Choose JSON objects randomly without repetition
    num_objects_to_choose = 4  # Specify the number of objects to choose
    random_objects = random.sample(data1, num_objects_to_choose)

    # Add chosen JSON objects to the list
    chosen_objects.extend(random_objects)
    
    
    return chosen_objects
    

#generate user email 
def generate_random_email(dealerEmail):
    # Generate a random username
    username = ''.join(random.choices(string.ascii_lowercase, k=8))

    # List of common email domains
    u, domain = dealerEmail.split("@")
    # Select a random domain from the list

    # Combine the username and domain to form the email address
    email = f"{username}@{domain}"
    return email

#generate user key 
def generate_user_secret(length=6):
    # Define the character pool for the secret
    characters = string.ascii_letters + string.digits + string.punctuation
    # Generate a random secret of the specified length
    secret = ''.join(secrets.choice(characters) for _ in range(length))
    salt = bcrypt.gensalt()
    hashed_password = bcrypt.hashpw(secret.encode('utf-8'), salt)
    json_serializable_data = hashed_password.decode('utf-8')
    return json_serializable_data  
    

for n in data:
    
    newD ={
        "dealerInfo" : {
            'name':n[0],
            'location':n[1],
            'phone':n[2],
            'mainEmail' : n[3], 
            'inventory':n[5],
            'services':n[6],
            'ratings':n[7], 
            'specialOffers':n[8],
            'hoursofOperation':n[9]
        },
        'credentialList':[
            {
                'workerEmail' : generate_random_email(n[3]),
                'workerHashedKey' : generate_user_secret()
            },
            {
                'workerEmail' : generate_random_email(n[3]),
                'workerHashedKey' : generate_user_secret()
            },
            {
                'workerEmail' : generate_random_email(n[3]),
                'workerHashedKey' : generate_user_secret()
            }
        ],
        'VehicleList': get_vhicleList()
    }
    

    dataJs.append(newD)
    
file_path = "dealerv1.json"
#save the new data to new json file 
with open(file_path, 'w') as f:
    json.dump(dataJs, f)
    
    
```
